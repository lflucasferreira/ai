# Aula 01 - Plots de Distribuições

```python
# Site: seaborn.pydata.org

# Dados: https://github.com/mwaskom/seaborn-data

import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=50) # quanto mais bins, mais espaçado
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # kind padrão scatter. Hex é outro tipo
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') # faz o jointplot de cada um dos dados numéricos. O hue faz a segregação
```

```python
sns.rugplot(tips['total_bill']) # utilidade de criar um KDE
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# Aula 02 - Plots Categóricos

```python
import numpy as np
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h') # só dados numéricos
```

```python
sns.violinplot(x='day', y='tip', data=tips) # modelo não paramétrico, KDE.
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True) # split junta os hues
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False) #padrão é True
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True) # split agora é dodge
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True) # recomendado para poucos dados
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar') # factorplor agora é catplot
```

```python
sns.clustermap(pf)
```

# Aula 04 - Plots de Regressão

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips) # linear model plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o', 'v'], scatter_kws={'s':100}) # s de size
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=0.6)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=0.6, height=8) # size agora é height
```

# Aula 05 - PairGrids

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist) # mapa da diagonal
g.map_upper(plt.scatter) # mapa na parte superior
g.map_lower(sns.kdeplot) # mapa na parte inferior
```

```python
sns.pairplot(iris, hue='species')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(plt.hist, 'total_bill')
```

# Aula 06 - Estilos e Cores

```python
tips.head()
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
sns.despine() # remove as linhas de cima e da direita
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
sns.despine(left=True) # remove as linhas de cima, da direita e da esquerda
```

```python
sns.set_style('ticks')
plt.figure(figsize=(12, 4)) # alterar tamanho do gráfico
sns.countplot(x='sex', data=tips)
```

```python
# https://matplotlib.org/tutorials/colors/colormaps.html
sns.set_context('poster', font_scale=0.6)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='inferno')
```
