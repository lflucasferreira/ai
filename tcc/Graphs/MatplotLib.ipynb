# Aula 01

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# funcional
plt.plot(x, y, 'r--') # ou color='r'
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1) # o último argumento é qual gráfico (elemento) estou querendo trabalhar, neste caso o primeiro
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [esquerda, baixo, largura, altura] -> 10% do canto esquerdo e inferior, e tem 80% de altura e largura
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('Título')
axes2.plot(y, x) # figura interna menor
```

# Aula 02

```python
fig, ax = plt.subplots() # o ax é um eixo único
ax.plot(x, x**3, 'b--') # método da instância de classe do eixo
ax.plot(x, x**4, 'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
```

```python
ax
```

```python
# posso iterar
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Título')
fig
```

```python
# ou posso acessar o elemento diretamente
ax[0].plot(x, y, 'b')
ax[0].set_title('Título')

ax[1].plot(x, y, 'g')
ax[1].set_title('Título')

fig
```

```python
fig, ax = plt.subplots(nrows = 5, ncols = 5)
plt.tight_layout() # para mostrar os gráficos sem sobreposição
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8, 5))

axes.plot(x, y, 'r-', label='x ˆ 2')
axes.plot(y, x, 'g--', label='x ** 1/2')
axes.set_title('Título')

axes.legend(loc=1) # obrigatório se informar os labels nos plots. O loc varia entre 0 e 10 (posições da legenda). O zero o matplotlib posiciona automaticamente onde for melhor
```

```python
fig.savefig('imagem.png') # salva o gráfico no formato png, mas poderia ser gif, jpg, etc
```

# Aula 03

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, x**2, 'g-.')
ax.plot(x, x**3, color='blue', linewidth=5, alpha=0.5, linestyle='--', marker='*') # alpha é a transparência, linestyle é o estilo da fonte (ls abreviado), linewidth pode ser abreviado para lw

ax.set_xlim([0, 2]) # intervalo do gráfico entre 0 e 2
ax.set_ylim([0, 10]) # pode mostrar retas em vezes de curva, para reduzir, aumentar o número do linspace.
```
